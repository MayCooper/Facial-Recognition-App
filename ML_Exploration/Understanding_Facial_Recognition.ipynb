# Understanding Facial Recognition

I'll employ key libraries such as numpy for array manipulation and mathematical operations, OpenCV for image processing, and matplotlib's pyplot for data visualization. 

I'll start by generating a sequential array of 100 elements, displaying its properties, and reshaping it into a matrix. Subsequently, a new array of the same shape is populated with random integers for comparative visualization.

The code further reads an image file in the BGR color scheme using OpenCV, and converts it into grayscale. 

After identifying the grayscale image's dimensions, the script extracts a 10x10 pixel slice and visualizes it using matplotlib. The script serves as an effective demonstration of array reshaping and grayscale image processing in Python.

```python
# Import the numpy library for handling arrays and numerical operations
import numpy as np 
# Import OpenCV (cv2) library for image processing tasks
import cv2 
# Import matplotlib's pyplot module for creating static, animated, and interactive visualizations
import matplotlib.pyplot as plt
# Set up matplotlib to work interactively in the notebook environment
%matplotlib inline
```

```python
# Create an array of numbers from 0 to 99
arr = np.arange(0,100,1)
```

```python
# Print the content of the array 'arr'
print(arr)
```

```python
# Get the total number of elements in the array 'arr'
len(arr)
```

```python
# Get the current shape of the array 'arr'
arr.shape
```

```python
# Reshape the array 'arr' into a 10x10 matrix
arr1 = arr.reshape((10,10))
```

```python
# Get the shape of the reshaped array 'arr1' (should be (10,10))
arr1.shape
```

```python
# Display the image of the reshaped array 'arr1' using a grayscale color map
plt.imshow(arr1,cmap='gray')
```

```python
# Create a new 10x10 array 'arr2' filled with random integers from 0 to 149
arr2 = np.random.randint(0,150,(10,10))
```

```python
# Show the randomly generated array 'arr2'
arr2
```

```python
# Display the image of the random array 'arr2' using a grayscale color map
plt.imshow(arr2,cmap='gray')
```

```python
# Read an image file 'test.jpg' from the directory 'data' (default color scheme is BGR)
img = cv2.imread('W:/MayCooperStation/New Documents/Data Science and ML/FacialRecognition/data/test.jpg')
```

```python
# Get the shape of the image 'img'
img.shape
```

```python
# Convert the BGR image 'img' to grayscale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
# Get the shape of the grayscale image 'gray'
gray.shape
```

```python
# Display the grayscale image 'gray' using a grayscale color map
plt.imshow(gray,cmap='gray')
```

```python
# Slice a 10x10 block from the top-left corner of the grayscale image 'gray'
slice = gray[0:10,0:10]
```

```python
# Display the sliced portion of the grayscale image using a grayscale color map
plt.imshow(slice,cmap='gray')
```
